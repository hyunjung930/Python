# DataFrame

```python
df = pd.DataFrame([["홍길동",90,80,70],["강감찬",80,78,90]])
print(df)
```

```python
nums=[1,2]
cnames = ["이름","국어","수학","영어"]
df = pd.DataFrame([["홍길동",90,80,70],["강감찬",80,78,90]],index=nums,columns=cnames)
print(df)
```

```python
table_data = {"연도":[2019,2020,2021],
              "매출":[100,130,120]}
df = pd.DataFrame(table_data)
print(df)
```

```python
#연산 More
s1 = pd.Series([1,2,3,4],index=['a','b','c','d'])
s2 = pd.Series([2,3,7,8],index=['a','c','d','e'])
s3 = s1+s2
print(s3)
```

```python
data1 = {"홍길동":[90,85,70],
        "강감찬":[99,80,65],
        "이순신":[100,100,100]}
df1 = pd.DataFrame(data1)
```

```python
data2 = {"홍길동":[100,75,80],
        "강감찬":[90,85,75],
        "을지문덕":[95,100,95]}
df2 = pd.DataFrame(data2)
```

```python
df3 = df1+df2
print(df3)
```
