# 행렬의 기본 연산

```python
import numpy as np
```

```python
ls1 = [1,2]
ls2 = [3,4]
ls3 = ls1 + ls2
print(ls3)
```

```python
s = 0
for i in range(10):
  s += i

assert s ==45,"s는 45여야 합니다." # assert 오류가 났는지 확인하기 위해 사용
# assert 조건, '메시지'로 사용
```

```python
#ls 목록에 있는 요소의 합계를 구하는 함수
def calsum(ls):
  s = 0
  for e in ls:
    s+=e
  return s
```

```python
assert calsum([1,2,3,2])==8,"calsum([1,2,3,2]) 값은 8이어야합니다."
```

```python
def add(l1,l2):
  assert len(l1) == len(l2), "더하기 연산을 위해서는 요소의 길이가 같아야 합니다."
# assert : 원하는 대로 작동되지 않을 때 오류를 내게 하는것 
  re =[]
  for i in range(len(l1)):
    re.append(l1[i]+l2[i])
  return re
```

```python
ls1 = [1,2]
ls2 = [3,4]
ls3 = add(ls1, ls2)
print(ls3)
```

```python
n1 = np.random.randint(1,1000,100,dtype=int)
# 1~1000 사이에 int 타입 100개 정수 무작위 추출
print(n1)
```

```python
n1 = np.array([[1,2],[3,4]])
n2 = np.array([4,5])
n3 = np.dot(n1,n2) # 행렬의 내적함수 dot을 이용 곱하기.
print(n3)
```

```python
n1 = np.array([1,2,3,4])
n2 = n1.reshape((2,2))
print(n2)
```

```python
n3 = np.transpose(n2) #전치행렬: 행렬의 대각선을 기준으로 반대로 교차 시켜 출력하는 것
print(n3)
```

```python
n3 = np.invert(n2) #invert: 반전(원소의 메모리)
print(n3)
```

```python
n3 = np.linalg.inv(n2) #역행렬
print(n3)
```

```python
n4 = n2.dot(n3)
print(n4)
```

```python
n2 = n1/3
print(n2)
```

# 통계관련

```python
n1 = np.random.randint(1,100,30,dtype = int)
print(n1)
```

```python
print("길이: ",len(n1))
```

```python
print("평균: ", n1.mean())
```

```python
n2 = n1 -n1.mean()
print("수평 이동한 후 평균:", n2.mean())
```

```python
print("최대값 위치:",n1.argmax())
print("최대값:",n1[16])
print("최대값:",n1.max())
```

```python
print("분산:",n1.var())
```

```python
print("표준편차:",n1.std())
```

```python
n2 = n1/n1.std()
print("스케일 조절한 후 표준 편차:",n2.std())
```

```python
print("최솟값: ",n1.min())
print("중앙값: ",np.median(n1))
print("1/4분위값: ",np.percentile(n1,0.25))
print("3/4분위값: ",np.percentile(n1,0.75))
```

```python
n1.sort()
print("누적 합:",np.cumsum(n1))
```

```python
n1 = np.arange(1,10)
print(n1)
print("누적 곱:",np.cumprod(n1))
```
