# 리스트 관련 함수

```python
ls = [10, 98, 11, 23, 45, 77, 6]
print("최솟값:", min(ls))
print("최대값:", max(ls))
print("합계: ", sum(ls))
```

```python
ls.sort()
print(ls)
ls.reverse()
print(ls)
```

```python
rls = reversed(ls)  # 변환 시킨것을 반환시켜줌
for elem in rls:
   print(elem, end =' ')
print()
print(ls)
```

```python
ls = [3,4,5,10,23,45,9]
els = enumerate(ls)
print(els)
```

```python
print(list(els))
```

```python
ls = [3,4,5,10,23,45,9]  # 리스트의 요소를 반복할 떄 인덱스가 몇번째인지 확이해야 하는 경우 enumerate 사용
for index, elem in enumerate(ls):
  print(f"{index}:{elem}")
```

# 딕셔너리

```python
ex_dict = {}
ex_dict["홍길동"] = 78
ex_dict["감길동"] = 78
ex_dict["고길동"] = 78
ex_dict["제갈길동"] = 78
print(ex_dict)
```

```python
print(ex_dict.items())
```

```python
for key, value in ex_dict.items():
 print(f"key:{key} value:{value}")
```

```python
sdict = sorted(ex_dict)
print(sdict)
```

# 고급진 리스트 컴프리핸션

1~ 100까지 구성한 리스트 만들기

```python
ls=[]
for i in range(100):
 ls.append(i)
print(ls)
```

```python
ls = [i for i in range(1,100)]
print(ls)
```

1에서 100사이의 숫자를 3배수로 리스트 표현

```python
ls = []
for i in range(100):
  if i % 3 == 0:
   ls.append(i)
print(ls)
```

```python
ls = [i for i in range(1,100) if i%3 ==0]   #list comprehensions
print(ls)
```
