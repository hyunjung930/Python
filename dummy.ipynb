```python
# 정상적으로 파일이 만들어졌는지 확인 하고 싶을 때
def print_n(s,n=1):
  for i in range(n):
    print(f"{i+1} s")
print_n("Hello",3) # print문 다같이 출력된다.
```

```python
def print_list(ls):
  for i,elem in enumerate(ls):
    print(f"{i+1} : {elem}")
```

```python
if __name__ == "__main__": # 현재 파일의 main이 진입점일 때 수행하시오.
  print_n("Hello",3)
  print_list([1,2,3])
```
