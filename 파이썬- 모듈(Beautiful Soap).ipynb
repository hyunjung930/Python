# 카카오 도서 검색 API 사용하기

쿼리 문자열 구조

도메인 주수? 매개변수 목록

매기변수 목록에서 매개변수와 매개변수&로 구분
매개변수는 키 = 값 구조를 지님

```python
# json 형식
import  urllib.request
url = 'https://dapi.kakao.com/v3/search/book'
target_param = "target=title"
page_param = "page=2"
query_in = input("검색 질의:")
query_param = "query="+urllib.parse.quote(query_in) # utf-8로 인코딩

query_str = url +'?'+target_param+'&'+query_param+'&'+page_param

print(query_str)

request = urllib.request.Request(query_str) # 서비스 요청에 사용할 개체를 생성
#request.add_header("Authorization", "KakaoAK 발급받은RESTAPI키")#발급받은RESTAPI키 부분은 자신에 맞게 수정하세요.
response = urllib.request.urlopen(request)
res = response.read().decode('utf-8') # utf-8 형식으로 불러옴 
print(res)
```

```python
import json
```

```python
jres = json.loads(res)
if jres == None:
  print("json,loads 실패")
  exit(0)
for post in jres['documents']:
  print("제목:", post['title'],"=====")
  print("설명:", post['contents'])
  print("url:", post['url'])
  print("가격:", post['price'])
  print("============================================")
```

# NAVER 도서 검색 API 사용하기

```python
# xml 형식
import  urllib.request
url = 'https://openapi.naver.com/v1/search/book.xml' # xml파일로
query_in = input("검색 질의:")
query_param = "query="+urllib.parse.quote(query_in) # utf-8로 인코딩

query_str = url +'?'+query_param

print(query_str)

request = urllib.request.Request(query_str) # 서비스 요청에 사용할 개체를 생성
#request.add_header('X-Naver-Client-Id','발급받은 client id')
#request.add_header('X-Naver-Client-Secret','발급받은 client secret')


response = urllib.request.urlopen(request)
res = response.read().decode('utf-8') # utf-8 형식으로 불러옴 
print(res)
```

```python
from bs4 import BeautifulSoup 
# bs4안에 있는 BeautifulSoup를 사용한다.
# passing 방법이 다르다/ xml은 BeautifulSoup 라이브러리를 가지고 사용
xmlsoup = BeautifulSoup(res,'html.parser')
items = xmlsoup.find_all('item') 
for item in items:
  print("제목: ", item.title.get_text()) # title이 태그 이름
  print("설명: ", item.description.get_text())
  print("가격: ", item.price.get_text())
  print("========================================================================")
```
